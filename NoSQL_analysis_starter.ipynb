```python
# Import dependencies
import pandas as pd
from pymongo import MongoClient
from pprint import pprint
import json
```

```python
FILE_PATH = "establishments.json"
PORT_NUMBER = 27017
DATA_BASE_NAME = "uk_food"
COLLECTION_NAME = "establishments"
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=PORT_NUMBER)
```

```python
# assign the uk_food database to a variable name
uk_food = mongo[DATA_BASE_NAME]
# assign the collection to a variable
collection = uk_food[COLLECTION_NAME]
```

```python
 # Loading or Opening the json file
with open(FILE_PATH, encoding="utf-8") as file:
    establishment_data = json.load(file)
```
